```python
# 敘述式練習

chinese = 78
english = 93
math = 73

total = chinese + english + math
print("total=", total)
```

```python
# 字串的相加
"total=" + "100"

# 字串 * 60遍
"=" * 60
```

```python
# 字串和變數的整合

apple = 5
orange = 6

print("蘋果=" + str(apple) + "個,橘子=" + str(orange) + "個")
```

```python
# str template
# 字串插補變數
# 字串和變數整合,傳出1個全新的字串
print(f"蘋果={apple}個,橘子={orange}個")
```

```python
# 內建函式 input(參數,引數值) 輸入
apple = int(input("請輸入蘋果的數量:"))
type(apple)
```
