```python
age = 19

# 雙項選擇
if age >= 18:   # ()可省略，條件>=18
    print("滿18了")
    print("可以考駕照")
else:   # 條件<18
    print("未滿18")
    print("不可以考駕照")
```

```python
#巢狀判斷
age = 50
if age < 18:
    print("未滿18")
    print("青少年")
else:
    if age < 40:
        print("青年")
    else:
        print("壯年")
```

```python
#多項選擇
age = 30
if age < 18:
    print("未滿18")
    print("青少年")
elif age < 40:
    print("青年")
else:
    print("壯年")
```

```python
score = 59
if score >= 90:
    print("優")
elif score >= 80:
    print("甲")
elif score >= 70:
    print("乙")
elif score >= 60:
    print("丙")
else:
    print("丁")
```

```python
#邏輯運算子寫法
score = 95
if score >= 90:
    print("優")
elif score< 90 and score >= 80:
    print("甲")
elif score < 80 and score >= 70:
    print("乙")
elif score < 70 and score >= 60:
    print("丙")
else:
    print("丁")
```

```python
# 僅Python特有寫法
score = 85
if score >= 90:
    print("優")
elif 90 > score >= 80:
    print("甲")
elif 80 > score >= 70:
    print("乙")
elif 70 > score >= 60:
    print("丙")
else:
    print("丁")
```

```python
# 數值與數學模組 random函式庫 
import random

while random.randint(1,10) != 5:
    print("這不是5")

print("這是5")
```

```python
import random

while True:
    rand_value = random.randint(1,10)
    if rand_value == 5:
        print("這是5")
        break
    print("這不是5")
```
